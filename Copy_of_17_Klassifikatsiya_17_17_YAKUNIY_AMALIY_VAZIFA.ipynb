**Mushroom dataset ustunlari va ularning tavsifi**  

1. **class** – qoʻziqorinning turi: zaharli (`p`) yoki zaharsiz (`e`). *Bu target (maqsad) ustun hisoblanadi.*  
2. **cap-shape** – qoʻziqorin shapkasi shakli:  
   - `b`: bell (qo‘ng‘iroq)  
   - `c`: conical (konussimon)  
   - `x`: convex (yumaloq)  
   - `f`: flat (tekis)  
   - `k`: knobbed (dumaloq tugunli)  
   - `s`: sunken (botiq)  

3. **cap-surface** – qoʻziqorin shapkasi yuzasi qanday koʻrinishda:  
   - `f`: fibrous (tolali)  
   - `g`: grooved (egik chiziqli)  
   - `y`: scaly (taroqli)  
   - `s`: smooth (silliq)  

4. **cap-color** – qoʻziqorin shapkasi rangi:  
   - `n`: brown (jigarrang)  
   - `b`: buff (to‘q sariq-jigarrang)  
   - `c`: cinnamon (dolchin rang)  
   - `g`: gray (kulrang)  
   - `r`: green (yashil)  
   - `p`: pink (pushti)  
   - `u`: purple (binafsha)  
   - `e`: red (qizil)  
   - `w`: white (oq)  
   - `y`: yellow (sariq)  

5. **bruises** – qoʻziqorin bosilganda dogʻ qoldiradimi?  
   - `t`: true (ha)  
   - `f`: false (yo‘q)  

6. **odor** – qoʻziqorin hidi:  
   - `a`: almond (bodom)  
   - `l`: anise (anis)  
   - `c`: creosote (kreozot – o‘ziga xos oʻtkir hid)  
   - `y`: fishy (baliq hidi)  
   - `f`: foul (yomon hid)  
   - `m`: musty (mog‘or hidi)  
   - `n`: none (hidsiz)  
   - `p`: pungent (keskin hid)  
   - `s`: spicy (achchiq hid)  

7. **gill-attachment** – qoʻziqorin plastinkalarining shapkaga bogʻlanish turi:

   - `d`: descending (pastga tushgan)

8. **gill-spacing** – plastinkalar orasidagi masofa:

11. **stalk-shape** – qoʻziqorin poyasi shakli:

12. **stalk-root** – qoʻziqorin poyasining ildiz qismi:

```python
# barcha kerakli kutubxonalar va toollarni import qilamiz
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn import metrics
from sklearn.metrics import  confusion_matrix, classification_report, accuracy_score
```

```python
url="https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/mushrooms.csv"
df = pd.read_csv(url)
df.head()
```

```python
# df o`lchami
df.shape
```

```python
# malumotlarni tekshiramiz va tahlil qilamiz

df.info()
```

```python
# nan qiymatlar mavjudligini tekshiramiz

df.isnull().sum()
```

```python
# barcha kategorikal ustunlarni numeric ko`rinishga o`tkazamiz

encoder = LabelEncoder()
for col in df.columns:
    df[col] = encoder.fit_transform(df[col])
```

```python
df.sample(10)
```

```python
# va bizda qaysi ustunlar class ustuniga kuchli bog`liqligini tekshiramiz

df.corrwith(df['class']).sort_values(ascending=False)
```

```python
# class ustuniga koproq bog`liq bo`lgan ustunlarni ajratib olamiz

df = df[['class','gill-size','population','habitat','stalk-surface-below-ring','gill-spacing','stalk-root','ring-type','bruises','gill-color']]
df.head(10)
```

```python
# bazi ustunlarni vizualizatsiya qilamiz

fig, axes = plt.subplots(1,4, figsize=(15,5))

# gill-size
sns.countplot(x="gill-size", data=df, ax=axes[0],hue='class')
axes[0].set_title("Gill-size")

#habitat
sns.countplot(x="habitat", data=df, ax=axes[1],hue='class')
axes[1].set_title("Habitat")

# Bruises
sns.countplot(x="bruises", data=df, ax=axes[2],hue='class')
axes[2].set_title("Bruises")

# gill-color
sns.countplot(x="gill-color", data=df, ax=axes[3],hue='class')
axes[3].set_title("Gill-color")

plt.tight_layout()

plt.show()
```

ML ga tayyorlaymiz

```python
df.head()
```

```python
# X va y ni ajratib olamiz
X = df.drop('class',axis=1)
y = df['class']
```

```python
# test va train qismlarga ajratamiz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y, random_state=42)
```

ML

K-NN

```python
# k-nn train qilamiz

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Modelni baholaymiz, f1, recall, precision
y_predict = knn.predict(X_test)
print(classification_report(y_test, y_predict))
print("Model aniqligi:", accuracy_score(y_test,y_predict))

# confusion matrix
sns.heatmap(confusion_matrix(y_test, y_predict), annot=True,fmt='g')
plt.show()
```

```python
# bar plot  report
report = classification_report(y_test, y_predict, output_dict=True)
df = pd.DataFrame(report).transpose()

df = df.iloc[:-3]

metrics = ["precision", "recall", "f1-score"]

x = np.arange(len(df.index))
width = 0.25

plt.figure(figsize=(10,6))

for i, metric in enumerate(metrics):
    plt.bar(x + i*width, df[metric], width, label=metric)

plt.xticks(x + width, df.index)
plt.ylabel("Score")
plt.title("Precision, Recall, F1-Score per Class")
plt.legend()

plt.show()
```

Logistic Regression

```python
# Modelni yaratamiz (training)
LR_model = LogisticRegression()
LR_model.fit(X_train, y_train)

# Modelni baholaymiz
y_predict = LR_model.predict(X_test)
print(classification_report(y_test, y_predict))
print("Model aniqligi:", accuracy_score(y_test,y_predict))

## confusion matrix
conf_mat = confusion_matrix(y_test, y_predict)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()
```

```python
# bar plot  report
report = classification_report(y_test, y_predict, output_dict=True)
df = pd.DataFrame(report).transpose()

df = df.iloc[:-3]

metrics = ["precision", "recall", "f1-score"]

x = np.arange(len(df.index))
width = 0.25

plt.figure(figsize=(10,6))

for i, metric in enumerate(metrics):
    plt.bar(x + i*width, df[metric], width, label=metric)

plt.xticks(x + width, df.index)
plt.ylabel("Score")
plt.title("Precision, Recall, F1-Score per Class")
plt.legend()

plt.show()
```

Decision tree

```python
# Modelni yaratamiz (training)
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)

# Modelni baholaymiz
y_predict = tree_model.predict(X_test)
print(classification_report(y_test, y_predict))
print("Model aniqligi:", accuracy_score(y_test,y_predict))

## confusion matrix
conf_mat = confusion_matrix(y_test, y_predict)
sns.heatmap(conf_mat, annot=True,fmt="g")
plt.show()
```
